# Synthetic Data Analysis
Visualizing and comparing real vs generated sensor data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
real_data = pd.read_csv('../data/raw/sensors_day1.csv')
generated_data = pd.read_csv('../data/synthetic/generated_batch_01.csv')
```

```python
real_data.hist(figsize=(12, 8))
plt.suptitle('Real Sensor Data')
plt.show()
```

```python
generated_data.hist(figsize=(12, 8))
plt.suptitle('Generated Sensor Data')
plt.show()
```
